```python
#count the number of restaurants/points of interest within each distance threshold.
within_quarter=df_all[df_all['miles'] <=.25]
within_half=df_all[df_all['miles'] <=.5]
within_one=df_all[df_all['miles'] <=1]
within_two=df_all[df_all['miles'] <=2]
within_three=df_all[df_all['miles'] <=3]
within_four=df_all[df_all['miles'] <=4]
within_five=df_all[df_all['miles'] <=5]
within_ten=df_all[df_all['miles'] <=10]

within_quarter
```

```python
#Groupby listing id. Gives count of points of interest within specified distance per airbnb listing, while preserving the neighborhood label.
within_quarter2=pd.DataFrame(within_quarter.groupby(['id', 'neighborhood_group'])['miles'].count()).reset_index()
within_half2=pd.DataFrame(within_half.groupby(['id', 'neighborhood_group'])['miles'].count()).reset_index()
within_one2=pd.DataFrame(within_one.groupby(['id', 'neighborhood_group'])['miles'].count()).reset_index()
within_two2=pd.DataFrame(within_two.groupby(['id', 'neighborhood_group'])['miles'].count()).reset_index()
within_three2=pd.DataFrame(within_three.groupby(['id', 'neighborhood_group'])['miles'].count()).reset_index()
within_four2=pd.DataFrame(within_four.groupby(['id', 'neighborhood_group'])['miles'].count()).reset_index()
within_five2=pd.DataFrame(within_five.groupby(['id', 'neighborhood_group'])['miles'].count()).reset_index()
within_ten2=pd.DataFrame(within_ten.groupby(['id', 'neighborhood_group'])['miles'].count()).reset_index()

within_five2
```

KMeans clustering.

KMeans is one of the simple but popular unsupervised learning algorithm. Here K indicates the number of clusters or classes the algorithm has to divide the data into. The algorithm starts with first group of randomly selected centroids, which are used as the beginning points for every cluster. It performs repetitive calculations to optimize the positions of the centroids.

For more information visit- https://en.wikipedia.org/wiki/K-means_clustering



```python
#Importing KMeans from sklearn

from sklearn.cluster import KMeans
```

```python
neighborhood_dist_to_poi_counts.columns
```

```python
# Now we calculate the Within Cluster Sum of Squared Errors (WSS) for different values of k. 
# Next, we choose the k for which WSS first starts to diminish.
X=neighborhood_dist_to_poi_counts[['quarter_mile', 'half_mile', 'one_mile', 'two_miles', 'three_miles', 'four_miles', 'five_miles', 'ten_miles']]
X
```

```python
wcss=[]

for i in range(1,10):
    km=KMeans(n_clusters=i)
    km.fit(X)
    wcss.append(km.inertia_)
```

```python
#The elbow curve

plt.figure(figsize=(12,6))

plt.plot(range(1,10),wcss)

plt.plot(range(1,10),wcss, linewidth=2, color="red", marker ="8")

plt.xlabel("K Value")
plt.xticks(np.arange(1,10,1))
plt.ylabel("WCSS")

plt.show()
```

```python
#this is known as the elbow graph , the x axis being the number of clusters
#the number of clusters is taken at the elbow joint point
#this point is the point where making clusters is most relevant
#the numbers of clusters is kept at maximum
```

```python
#Taking 3 clusters

km1=KMeans(n_clusters=3, random_state=44)
```

```python
#Fitting the input data

km1.fit(X)
```

```python
#predicting the labels of the input data

y=km1.predict(X)
```

```python
#adding the labels to a column named label

neighborhood_dist_to_poi_counts["label"] = y
```

```python
neighborhood_dist_to_poi_counts
```

```python
#The new dataframe with the clustering done

neighborhood_dist_to_poi_counts['label'] = neighborhood_dist_to_poi_counts['label'].replace([1,2,0],['Fewest Nearby Choices','Moderate Nearby Choices','Most Nearby Choices'])
```

```python
#Scatterplot of the clusters

plt.figure(figsize=(10,6))
sns.scatterplot(x = 'five_miles',y = 'neighborhood_group',hue="label",
                 palette=['orange','dodgerblue','green'], legend="full",data = neighborhood_dist_to_poi_counts,s = 60 )
plt.xlabel('five_miles')
plt.ylabel('neighborhood_group')
plt.title('Mean Average Number of Points of Interest Within 5 Miles of Each Airbnb, by Neighborhood')
plt.show()
```
